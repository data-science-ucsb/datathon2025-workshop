# What is Exploratory Data Analysis?

As the name suggests, <ins>Exploratory</ins> Data Analysis (EDA) is all about exploring data. **It encompasses everything you do before modeling.**

Typically, there are 4 components of an EDA, which can occur in any order.
- Asking questions
- Collecting data
- Cleaning data
- Understanding data

The goals of an EDA are to
1. Formulate appropriate questions for your data
2. Reveal patterns, strengths, and weaknesses of your data
3. Suggest reasonable models or assumptions for your data

Much of an analyst's work falls under EDA, so a good EDA is crucial to your toolbox.

## Cleaning Data

*What <ins>good</ins> is data science if everything is a mess?*

To clean data is to simply restore its shortcomings, remove its errors, or make it easier to work with. Finding a perfect dataset is near impossible, and there will always be some issue. You often identify these as you collect or explore data.

Bad data typically (but not always) comes in the form of **missing data**. [This official Panda's guide](https://pandas.pydata.org/docs/user_guide/missing_data.html) explains how to handle missing data very well. However, some tips to keep in mind

- [Dropping data](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.dropna.html#pandas.DataFrame.dropna) is the easiest option, but before doing so ask yourself the following
  1. How much data will I lose?
  2. Is there a pattern behind the missing data, or do things seem to be missing for no reason?
    - *Maybe the lack of data is data in and of itself*
  3. How valuable are the individual datapoints?
- [Filling data](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.fillna.html#pandas.DataFrame.fillna) is the most popular because it preserves data you would otherwise lose, but ask yourself
  1. How exactly are you going to fill missing values?
    - Filling values is going to introduce bias, so be sure to justify what you're filling with
  2. To what extent are you introducing bias? What is being affected?
  3. Could there be a pattern to the missingness that you're neglecting?

<a name="understand"></a>
## Understanding Data

*What good is <ins>data science</ins> if we don't even know what we're looking at?*

It's not enough to grab a dataset and throw arbitrary models onto it, this is how bad models are made! Rather, it's very important to understand the needs and workings of your data

How is this done? It depends what kind of data you're dealing with, but typically:

- Computing [summary statistics](https://www.cuemath.com/data/summary-statistics/#what-is-a-venn-diagram)
  - Mean, median, mode, standard deviation, maximum, minimum, etc.
  - *This only makes sense for numerical data.*
  - Read [this Panda's guide](https://pandas.pydata.org/docs/getting_started/intro_tutorials/06_calculate_statistics.html) for more
- Transforming your data
  - Infer new variables
    - *E.g. Convert time elapsed into time-of-day*
  - Normalize/standardize variables
    - *E.g. Adjusting prices for inflation*
  - Refer to the [Panda's documentation](https://pandas.pydata.org/docs/user_guide/index.html) for available transformations
- **Visualizing your data**
  - *Typically the most insightful*
  - Scatterplots, histograms, distribution plots, violin plots, etc.
  - Read [this Seaborn guide](https://seaborn.pydata.org/tutorial/introduction.html) for more info

The goal of understanding your data should be to
1. Refine your questions to be more in tune with your data
2. Learn the strengths and weaknesses of your data
3. Identify new variables to consider
4. Gauge reasonable assumptions/models for your data

# A Second Example

To communicate both the importance of an EDA and (roughly) what it looks like, we'll be looking at a dataset known as the [Datasaurus Dozen](https://www.kaggle.com/code/tombutton/datasaurus-dozen).

```python
!git clone https://github.com/DanLeEpicMan/datasauras-dozen.git

data = pd.read_csv('datasauras-dozen/datasaurus.csv')
```

The dataset is structured as a 2D point (`x, y` columns) with a label (`dataset` column). Due to the nature of this example, I will primarily focus on exploring and understanding this dataset.

## Summarizing the Data

Let's just take a glance at the summary statistics.

```python
# if you only want specific summary statistics,
# look into https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.agg.html#pandas.DataFrame.agg
data.describe()
```

And now let's look at the summary statistics across each category.

```python
# see groupby: https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.groupby.html
# this essentially allows us to split our dataframes into several dataframes
# so we're grouping data by its value in the 'dataset' column

data.groupby('dataset').describe()
```

Notice how **the mean, count, and standard deviations** are all almost exactly the same, both in aggregate and across each category. Moreover, let's graph all our data together

```python
# see jointplot: https://seaborn.pydata.org/generated/seaborn.jointplot.html
# as opposed to scatterplot: https://seaborn.pydata.org/generated/seaborn.scatterplot.html

# primary difference is jointplot also plots marginal distributions of each axis,
# whereas scatterplot let's you make more refined scatterplots

sns.jointplot(data=data, x='x', y='y')
```

At a glance it seems like a circular model would be appropriate. Perhaps you might consider fitting a multivariate normal distribution.

## Looking at this dataset closer...

Let's now plot a scatterplot for each category

```python
sns.relplot(data=data, x='x', y='y', col='dataset', col_wrap=4);
```

Notice how these graphs are incredibly different from each other! Yet all our previous analysis concluced that the categories are more or less the same across the board and can be pooled together. Clearly, this shows our initial impressions were wrong. **In particular, this demonstrates that looking at summary statistics and a single scatterplot alone wasn't enough to understand our data.**
